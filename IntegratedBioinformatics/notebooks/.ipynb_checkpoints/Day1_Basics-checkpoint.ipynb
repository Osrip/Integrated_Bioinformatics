```python
# preferably add a code block for each operation (e.g. by selecting a cell and pressing B on the keyboard)
index = np.arange(1,6)
df = pd.DataFrame({'a':np.arange(2,7),
                  'b':np.array([3,1,9,12,2]),
                   'c':np.array([5.,1.4,1.2,0.8,2.3]),
                   'd':np.array(['horse', 'cat', 'dog','tortoise', 'rabbit']),
                   'e':np.array([13,11,np.NaN ,np.NaN ,9])}, index = index)
df
```

# Data Merging

* Merge df3 with df4 on columns c, d and e. Change columns with same names in a meaningful way.

```python
print(df3)
print(df4)
def add_to_col_name(df, add_str, except_col):
    #renames columns
    col = df.columns
    #new_col = [en + add_str for en in col]
    new_col = []
    for en in col:
        if en != except_col:
            new_col.append(en + add_str)
        else :
            new_col.append(en)
    df.columns = new_col
    
add_to_col_name(df3, '3', 'd')
add_to_col_name(df4, '4', 'd')

df4.merge(df3, left_on='d', right_on='d')
```

* Now perform a full outer join of the data, to keep all rows contained in either of the two dataframes.

```python
#reloading d3 and d4
df4 = pd.read_csv('../example_data/df4.csv')
df4
df3 = pd.read_csv('../example_data/example_df2.csv', sep = ';')
df3

add_to_col_name(df3, '3', 'd')
add_to_col_name(df4, '4', 'd')

df4.merge(df3, on = 'd', how = 'outer')
```
